```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder \
    .appName("Metroid Analysis") \
    .getOrCreate()
```

```
df = spark.read.option("multiline","true").json("s3://checkingacc/metroids.json")
```

```
df = df.drop(':@computed_region_cbhk_fwbd', ':@computed_region_nnqa_25f4')
#auto-generated by Spark when the JSON data contains complex nested structures or arrays
```

```
display(df)
```

```
df.createOrReplaceTempView("temp_table")
```

```
df_geolocation = spark.sql("""
    SELECT 
        geolocation.coordinates[0] AS longitude,
        geolocation.coordinates[1] AS latitude,
        geolocation.type AS geo_type
    FROM temp_table
""")
```

```
display(df_geolocation)
```

```
df_metroid = df.drop("geolocation")
```

```
display(df_metroid)
```

```
df_metroid.createOrReplaceTempView("metroid_table")
```

```
df_publish = spark.sql("""
    SELECT id, name, nametype, recclass, fall,
        year, reclat, reclong,
        mass,
        (mass - LAG(mass, 1, mass) OVER (PARTITION BY year ORDER BY year)) / LAG(mass, 1, mass) OVER (PARTITION BY year ORDER BY year) * 100 AS percentage_growth
    FROM metroid_table
""")
```

```
display(df_publish)
```

```
import pandas as pd
```

```
df_pandas = df.toPandas()
```

```
recclass_trend_query = """
    SELECT year, recclass, COUNT(*) AS count
    FROM publish
    GROUP BY year, recclass
    ORDER BY year, recclass
"""
```

```
df_publish.createOrReplaceTempView("publish")
```

```
recclass_trend_data = spark.sql(recclass_trend_query).toPandas()
```

```
mass_growth_query = """
    SELECT year, SUM(percentage_growth) AS sum_percentage_growth
    FROM publish
    GROUP BY year
    ORDER BY year
"""
```

```
mass_growth_data = spark.sql(mass_growth_query).toPandas()
```

```
import matplotlib.pyplot as plt
```

```
mass_growth_data['year'] = mass_growth_data['year'].str[:4]
```

```
mass_growth_data['year'] = pd.to_numeric(mass_growth_data['year'], errors='coerce')
```

```
display(mass_growth_data)
```

```
mass_growth_data = mass_growth_data.dropna(subset=['year'])
```

```
mass_growth_data['year'] = mass_growth_data['year'].astype(int)
```

```
# Percentage growth of 'mass' for each year
plt.figure(figsize=(12, 6))
plt.plot(mass_growth_data['year'], mass_growth_data['sum_percentage_growth'], marker='o')
plt.xlabel('Year')
plt.ylabel('Percentage Growth of Mass')
plt.title('Percentage Growth of Mass for Each Year')
plt.grid(True)
plt.show()
```

```
recclass_trend_data['year'] = recclass_trend_data['year'].str[:4]
```

```
recclass_trend_data['year'] = pd.to_numeric(recclass_trend_data['year'], errors='coerce')
```

```
display(recclass_trend_data)
```
